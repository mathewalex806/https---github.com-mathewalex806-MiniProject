```python
import os
import json
import shutil
import random

# Set the root directories for labels and source images.
labels_root = '/content/sen12flood/sen12flood/sen12floods_s1_labels/sen12floods_s1_labels'
# Adjust the source_root according to your dataset structure.
source_root = '/content/sen12flood/sen12flood/sen12floods_s1_source/sen12floods_s1_source'

# Counters for flooding True/False
flooding_true_count = 0
flooding_false_count = 0

# List to hold names of directories where FLOODING is False.
false_dirs = []

# Iterate through each subdirectory in the labels root directory.
for dir_name in os.listdir(labels_root):
    dir_path = os.path.join(labels_root, dir_name)
    if os.path.isdir(dir_path):
        labels_geojson_path = os.path.join(dir_path, 'labels.geojson')
        if os.path.exists(labels_geojson_path):
            with open(labels_geojson_path, 'r') as f:
                data = json.load(f)
                flooding = data.get("properties", {}).get("FLOODING")
                if flooding is True:
                    flooding_true_count += 1
                elif flooding is False:
                    flooding_false_count += 1
                    false_dirs.append(dir_name)
        else:
            print(f"File not found: {labels_geojson_path}")

print("Total labels with flooding true:", flooding_true_count)
print("Total labels with flooding false:", flooding_false_count)

# Calculate how many false directories to delete to achieve a 50-50 balance.
num_to_delete = flooding_false_count - flooding_true_count
print("Deleting", num_to_delete, "false label directories to balance the dataset.")

# Randomly select directories to delete.
dirs_to_delete = random.sample(false_dirs, num_to_delete)

# Delete each selected label directory and its corresponding source directory.
for dir_name in dirs_to_delete:
    label_dir = os.path.join(labels_root, dir_name)
    source_dir = os.path.join(source_root, dir_name)

    if os.path.exists(label_dir):
        shutil.rmtree(label_dir)
        print(f"Deleted label directory: {label_dir}")
    else:
        print(f"Label directory not found: {label_dir}")

    if os.path.exists(source_dir):
        shutil.rmtree(source_dir)
        print(f"Deleted source directory: {source_dir}")
    else:
        print(f"Source directory not found: {source_dir}")
```

```python
import os
import torch
import numpy as np
import rasterio
import json
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F

class FloodDataset(Dataset):
    def __init__(self, source_dir, labels_dir, target_size=(512, 512), transform=None):
        self.source_dir = source_dir
        self.labels_dir = labels_dir
        self.target_size = target_size  # Fixed size for all images
        self.transform = transform

        # Get list of SAR image directories (sorted for consistency)
        self.sample_ids = sorted(os.listdir(source_dir))

    def __len__(self):
        return len(self.sample_ids)

    def preprocess_sar(self, image):
        """Apply log transformation and normalize to [-1,1]."""
        # Apply log transformation first
        image = 10 * np.log10(image + 1e-6) + 50  # Avoid log(0)

        # Compute denominator safely
        img_min = np.min(image)
        img_max = np.max(image)
        denom = img_max - img_min
        if denom < 1e-6:
            # If the image has no contrast, set it to zeros
            image = np.zeros_like(image)
        else:
            image = (image - img_min) / denom  # Normalize to [0,1]

        # Scale to [-1,1]
        image = (image * 2) - 1
        return image.astype(np.float32)

    def get_flood_label(self, label_path):
        """Read labels.geojson and return flood label (0 or 1)."""
        if not os.path.exists(label_path):
            print(f"Warning: Missing label file {label_path}")
            return 0  # Default to no flood if label is missing

        with open(label_path, "r") as f:
            label_data = json.load(f)
        return 1 if label_data["properties"].get("FLOODING", False) else 0

    def __getitem__(self, idx):
        sample_id = self.sample_ids[idx]
        sample_dir = os.path.join(self.source_dir, sample_id)

        # Find VH.tif within the sample directory (search recursively)
        vh_file = None
        for root, dirs, files in os.walk(sample_dir):
            if "VH.tif" in files:
                vh_file = os.path.join(root, "VH.tif")
                break
        if vh_file is None:
            raise FileNotFoundError(f"VH.tif not found in directory: {sample_dir}")
        image_path = vh_file

        # Construct label path.
        # Assuming the folder names differ only by replacing "sen12floods_s1_source" with "sen12floods_s1_labels".
        label_subdir = sample_id.replace("sen12floods_s1_source", "sen12floods_s1_labels")
        label_path = os.path.join(self.labels_dir, label_subdir, "labels.geojson")

        # Check if label file exists
        if not os.path.exists(label_path):
            print(f"Warning: Missing label for {sample_id}, skipping.")
            return self.__getitem__((idx + 1) % len(self.sample_ids))

        # Load SAR image
        with rasterio.open(image_path) as src:
            image = src.read(1).astype(np.float32)  # Read single-band VH SAR image
        image = self.preprocess_sar(image)

        # Convert the numpy array to a torch tensor and add channel dimension: (1, H, W)
        image = torch.tensor(image).unsqueeze(0)
        # Resize image to target size (batch stacking requires same size)
        image = F.interpolate(image.unsqueeze(0), size=self.target_size, mode='bilinear', align_corners=False).squeeze(0)

        # Load flood label (global scalar label: 0 or 1)
        flood_label = self.get_flood_label(label_path)
        label = torch.tensor(flood_label, dtype=torch.float32)

        return image, label

# Update the paths for Colab.
# Here, we assume that you have unzipped your dataset in /content/sen12flood/
dataset = FloodDataset(
    source_dir="/content/sen12flood/sen12flood/sen12floods_s1_source/sen12floods_s1_source",
    labels_dir="/content/sen12flood/sen12flood/sen12floods_s1_labels/sen12floods_s1_labels",
    target_size=(512, 512)
)
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# Check a sample:
sample_image, sample_label = dataset[0]
print("Image shape:", sample_image.shape)  # Expected: (1, 512, 512)
print("Flood label:", sample_label)  # 0 or 1
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class UNet(nn.Module):
    def __init__(self, in_channels=1, out_channels=1):
        super(UNet, self).__init__()

        # Encoder
        self.enc1 = self.double_conv(in_channels, 64)
        self.enc2 = self.double_conv(64, 128)
        self.enc3 = self.double_conv(128, 256)
        self.enc4 = self.double_conv(256, 512)

        # Bottleneck
        self.bottleneck = self.double_conv(512, 1024)

        # Decoder
        self.upconv4 = self.upconv(1024, 512)
        self.dec4 = self.double_conv(1024, 512)
        self.upconv3 = self.upconv(512, 256)
        self.dec3 = self.double_conv(512, 256)
        self.upconv2 = self.upconv(256, 128)
        self.dec2 = self.double_conv(256, 128)
        self.upconv1 = self.upconv(128, 64)
        self.dec1 = self.double_conv(128, 64)

        # Output layer
        self.final_conv = nn.Conv2d(64, out_channels, kernel_size=1)

    def forward(self, x):
        # Encoder
        e1 = self.enc1(x)
        e2 = self.enc2(self.downsample(e1))
        e3 = self.enc3(self.downsample(e2))
        e4 = self.enc4(self.downsample(e3))

        # Bottleneck
        b = self.bottleneck(self.downsample(e4))

        # Decoder
        d4 = self.upconv4(b)
        d4 = torch.cat([d4, e4], dim=1)
        d4 = self.dec4(d4)

        d3 = self.upconv3(d4)
        d3 = torch.cat([d3, e3], dim=1)
        d3 = self.dec3(d3)

        d2 = self.upconv2(d3)
        d2 = torch.cat([d2, e2], dim=1)
        d2 = self.dec2(d2)

        d1 = self.upconv1(d2)
        d1 = torch.cat([d1, e1], dim=1)
        d1 = self.dec1(d1)

        # Output
        return torch.sigmoid(self.final_conv(d1))

    def double_conv(self, in_channels, out_channels):
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
        )

    def downsample(self, x):
        return F.max_pool2d(x, kernel_size=2, stride=2)

    def upconv(self, in_channels, out_channels):
        return nn.ConvTranspose2d(in_channels, out_channels, kernel_size=2, stride=2)

# Set up device for GPU (T4 in Colab)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Initialize the model and move it to the GPU
model = UNet(in_channels=1, out_channels=1).to(device)
print(model)

# Optionally, test with a dummy input to ensure everything runs on GPU
dummy_input = torch.randn(1, 1, 512, 512).to(device)
output = model(dummy_input)
print("Output shape:", output.shape)
```

```python
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from tqdm.notebook import tqdm  # Use notebook version for better progress display in Colab

# Device configuration: use GPU if available (T4 in Colab)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Assume UNet is defined in a previous cell; if not, define it here.
# For example, if you have the UNet class defined, you can initialize it as follows:
model = UNet(in_channels=1, out_channels=1).to(device)

# Define the loss function and optimizer
criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# Create the DataLoader.
# Note: 'dataset' should be defined previously. Make sure your dataset is stored in Colab's file system,
# either by mounting your Google Drive or uploading the dataset.
dataloader = DataLoader(dataset, batch_size=8, shuffle=True, num_workers=2, pin_memory=True)

num_epochs = 100
total_start_time = time.time()

for epoch in range(num_epochs):
    epoch_start = time.time()
    model.train()
    running_loss = 0.0

    for images, labels in tqdm(dataloader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        images = images.to(device)
        # Adjust labels shape to (batch, 1)
        labels = labels.to(device).unsqueeze(1)

        optimizer.zero_grad()

        # Forward pass: model outputs shape (batch, 1, H, W)
        outputs = model(images)

        # Global average pooling to get a single prediction per image (output shape: (batch, 1))
        outputs = torch.mean(outputs, dim=(2, 3))

        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

    epoch_time = time.time() - epoch_start
    epoch_loss = running_loss / len(dataset)
    remaining_epochs = num_epochs - epoch - 1
    estimated_remaining_time = remaining_epochs * epoch_time
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}, Epoch Time: {epoch_time:.2f} sec, "
          f"Estimated Remaining Time: {estimated_remaining_time:.2f} sec")

total_time = time.time() - total_start_time
print(f"Training complete in {total_time/60:.2f} minutes.")

# Save the model to the current working directory in Colab
torch.save(model.state_dict(), "unet_flood_model.pth")
print("Model saved as unet_flood_model.pth")
```

```python
import os
import json
import shutil

def delete_invalid_pairs(source_dir, labels_dir):
    """
    Checks each sample directory in source_dir and verifies that the corresponding label directory
    contains a valid stac.json file with a 'labels' asset. If not, deletes both the source sample
    directory and the label directory.

    Assumes the sample directory name in the source contains 'sen12floods_s1_source'
    which is replaced by 'sen12floods_s1_labels' to form the label directory name.

    Returns the number of invalid sample pairs deleted.
    """
    invalid_count = 0
    samples = os.listdir(source_dir)

    for sample in samples:
        sample_path = os.path.join(source_dir, sample)
        if os.path.isdir(sample_path):
            # Construct corresponding label directory path.
            label_subdir = sample.replace("sen12floods_s1_source", "sen12floods_s1_labels")
            label_dir = os.path.join(labels_dir, label_subdir)
            stac_file = os.path.join(label_dir, "stac.json")

            valid_label = False
            if os.path.exists(stac_file):
                try:
                    with open(stac_file, 'r') as f:
                        stac_data = json.load(f)
                    # Check that the 'assets' key exists and includes the 'labels' asset.
                    if "assets" in stac_data and "labels" in stac_data["assets"]:
                        valid_label = True
                except Exception as e:
                    print(f"Error reading {stac_file}: {e}")
                    valid_label = False
            else:
                valid_label = False

            if not valid_label:
                invalid_count += 1
                print(f"Deleting invalid pair: {sample} has no valid label asset.")
                # Delete the source sample directory
                if os.path.exists(sample_path):
                    shutil.rmtree(sample_path)
                    print(f"Deleted source directory: {sample_path}")
                else:
                    print(f"Source directory not found: {sample_path}")

                # Delete the corresponding label directory if it exists
                if os.path.exists(label_dir):
                    shutil.rmtree(label_dir)
                    print(f"Deleted label directory: {label_dir}")
                else:
                    print(f"Label directory not found: {label_dir}")

    print(f"Total invalid sample pairs deleted: {invalid_count}")
    return invalid_count

# Example usage:
source_directory = "/content/sen12flood/sen12flood/sen12floods_s1_source/sen12floods_s1_source"
labels_directory = "/content/sen12flood/sen12flood/sen12floods_s1_labels/sen12floods_s1_labels"

deleted_pairs = delete_invalid_pairs(source_directory, labels_directory)
print(f"Total pairs deleted: {deleted_pairs}")
```
